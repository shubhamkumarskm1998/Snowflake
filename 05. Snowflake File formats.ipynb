# ❄️ Snowflake — File Format vs COPY Command (Detailed Notes + Examples)

---

# 🧠 1. What is File Format in Snowflake?

👉 A File Format defines how Snowflake reads and interprets data files

---

## 💡 Simple Idea

👉 File Format = rules to read file

- Delimiter
- Header
- Encoding
- File type

---

## Supported Formats

- CSV
- JSON
- Parquet
- Avro
- ORC

---

## Example

CSV file:

id,name,amount
1,A,100
2,B,200

---

# ⚙️ 2. Creating File Format

## CSV

CREATE FILE FORMAT my_csv_format
TYPE = 'CSV'
FIELD_DELIMITER = ','
SKIP_HEADER = 1;

## JSON

CREATE FILE FORMAT my_json_format
TYPE = 'JSON';

## PARQUET

CREATE FILE FORMAT my_parquet_format
TYPE = 'PARQUET';

---

# 🧠 3. What is COPY INTO?

👉 COPY INTO is used to load or unload data

---

# ⚙️ 4. Load Data

COPY INTO orders
FROM @my_stage
FILE_FORMAT = my_csv_format;

---

# 📤 5. Unload Data

COPY INTO @my_stage
FROM orders
FILE_FORMAT = (TYPE = 'CSV');

---

# 🔄 6. Difference

File Format → defines structure
COPY INTO → moves data

---

# 🧠 7. End-to-End Example

CREATE TABLE orders (
    id INT,
    name STRING,
    amount NUMBER
);

CREATE FILE FORMAT my_csv_format
TYPE = 'CSV'
FIELD_DELIMITER = ','
SKIP_HEADER = 1;

PUT file://orders.csv @my_stage;

COPY INTO orders
FROM @my_stage
FILE_FORMAT = my_csv_format;

SELECT * FROM orders;

---

# 🧠 8. Inline vs Named

Inline:

FILE_FORMAT = (TYPE='CSV' SKIP_HEADER=1)

Named:

FILE_FORMAT = my_csv_format

---

# 🚀 One Line

File Format = rules
COPY = action
