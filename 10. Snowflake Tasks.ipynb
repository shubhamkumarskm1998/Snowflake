# ❄️ Snowflake Tasks — Detailed Notes with Examples

---

# 🧠 1. What is a Task?

👉 A **Task** in Snowflake is used to **schedule and automate SQL operations**

---

## 💡 Simple Idea

👉 Task = **Scheduler (like cron job)**

---

## 🎯 Purpose

- Automate pipelines
- Run queries on schedule
- Work with Streams for real-time processing

---

# 📊 2. Types of Tasks

---

## 🔹 1. Scheduled Task

👉 Runs on a fixed time schedule

### Example:

CREATE TASK my_task
WAREHOUSE = my_wh
SCHEDULE = '1 MINUTE'
AS
INSERT INTO target_table
SELECT * FROM source_table;

---

## 🔹 2. Triggered Task (Stream-based)

👉 Runs only when data changes

### Example:

CREATE TASK stream_task
WAREHOUSE = my_wh
WHEN SYSTEM$STREAM_HAS_DATA('orders_stream')
AS
INSERT INTO target_table
SELECT * FROM orders_stream;

---

## 🔹 3. DAG / Dependent Tasks

👉 Tasks can depend on other tasks

### Example:

CREATE TASK task2
AFTER task1
AS
INSERT INTO final_table SELECT * FROM temp_table;

---

# ⚙️ 3. Scheduling Methods

---

## ⏰ 1. Time-based Scheduling

- Minutes
- Hours
- Cron expressions

### Example:

SCHEDULE = '5 MINUTE'

---

### CRON Example:

SCHEDULE = 'USING CRON 0 * * * * UTC'

👉 Runs every hour

---

## ⚡ 2. Event-based Scheduling

👉 Using Streams

WHEN SYSTEM$STREAM_HAS_DATA('stream_name')

---

# 🧠 4. Manual vs Task-Based Execution

---

## ❌ Manual Approach

```sql
INSERT INTO target_table
SELECT * FROM source_table;
```

Problems:
- Manual effort
- Not scalable
- Risk of missing runs

---

## ✅ Task-Based Approach

```sql
CREATE TASK auto_task
WAREHOUSE = my_wh
SCHEDULE = '1 MINUTE'
AS
INSERT INTO target_table
SELECT * FROM source_table;
```

Benefits:
- Automated
- Reliable
- Scalable

---

# 🔄 5. Resume & Suspend Tasks

---

## ▶️ Resume Task

```sql
ALTER TASK my_task RESUME;
```

👉 Starts execution

---

## ⏸️ Suspend Task

```sql
ALTER TASK my_task SUSPEND;
```

👉 Stops execution

---

## 📌 Important

- Tasks are **created in suspended state by default**

---

# 🔗 6. Using Tasks with Streams (Real-Time)

---

## Example Pipeline

Step 1: Create Stream

```sql
CREATE STREAM orders_stream ON TABLE orders;
```

---

Step 2: Create Task

```sql
CREATE TASK process_orders
WAREHOUSE = my_wh
SCHEDULE = '1 MINUTE'
AS
INSERT INTO analytics_table
SELECT * FROM orders_stream;
```

---

Step 3: Resume Task

```sql
ALTER TASK process_orders RESUME;
```

---

## Flow

App → Table → Stream → Task → Target Table

---

# 🚀 7. Real Production Example

---

## Scenario: E-commerce

- Orders coming continuously
- Need analytics in near real-time

---

## Implementation

```sql
CREATE STREAM orders_stream ON TABLE orders;

CREATE TASK process_orders
WAREHOUSE = my_wh
WHEN SYSTEM$STREAM_HAS_DATA('orders_stream')
AS
INSERT INTO analytics_orders
SELECT * FROM orders_stream;

ALTER TASK process_orders RESUME;
```

---

## Result

- No manual work
- Real-time updates
- Efficient pipeline

---

# ⚡ 8. Best Practices

- Use **Streams + Tasks together**
- Keep tasks lightweight
- Use **cron for production scheduling**
- Monitor using TASK_HISTORY

# 🚨 10. Important Points

- Task = scheduler
- Default state = suspended
- Needs warehouse (except serverless tasks)
- Can create dependency chain
