# ❄️ Snowflake — Stages (Internal, External, User) + UI, SQL & Python Examples

---

# 🧠 1. What is a Stage in Snowflake?

👉 A stage is a location where files are stored (for loading/unloading data)

---

## 💡 Simple Idea

👉 Stage = temporary storage area for files

- CSV
- JSON
- Parquet

---

## 🎯 Why Stages?

- Load data into Snowflake
- Export data
- Integrate with data lakes

---

# 📊 2. Types of Stages

1. Internal Stage
2. External Stage
3. User Stage

---

# 🟢 3. Internal Stage

## Example (SQL)

CREATE STAGE my_stage;

PUT file://data.csv @my_stage;

COPY INTO orders
FROM @my_stage
FILE_FORMAT = (TYPE = 'CSV');

---

# 🔵 4. External Stage

## Example

CREATE STAGE ext_stage
URL='s3://my-bucket/data/'
CREDENTIALS=(AWS_KEY_ID='xxx' AWS_SECRET_KEY='xxx');

COPY INTO orders
FROM @ext_stage
FILE_FORMAT = (TYPE = 'CSV');

---

# 🟣 5. User Stage

## Syntax

@~

## Example

PUT file://data.csv @~;

COPY INTO orders
FROM @~
FILE_FORMAT = (TYPE = 'CSV');

---

# 🧱 6. Table Stage

## Syntax

@%table_name

## Example

PUT file://data.csv @%orders;

COPY INTO orders FROM @%orders;

---

# ⚙️ 9. Using via SQL

CREATE TABLE orders (
    id INT,
    amount NUMBER
);

CREATE STAGE my_stage;

PUT file://orders.csv @my_stage;

COPY INTO orders
FROM @my_stage
FILE_FORMAT = (TYPE = 'CSV' SKIP_HEADER = 1);

---

# 🐍 10. Using via Python

import snowflake.connector

conn = snowflake.connector.connect(
    user='USER',
    password='PASSWORD',
    account='ACCOUNT'
)

cur = conn.cursor()

cur.execute("CREATE STAGE my_stage")
cur.execute("PUT file://orders.csv @my_stage")
cur.execute("""
COPY INTO orders
FROM @my_stage
FILE_FORMAT = (TYPE = 'CSV' SKIP_HEADER = 1)
""")

cur.close()
conn.close()
