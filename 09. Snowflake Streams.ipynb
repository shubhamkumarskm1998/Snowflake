# ❄️ Snowflake Streams — Complete Detailed Notes (Types, Usage, Deployment, Real-World)

---

# 🧠 1. What is a Stream in Snowflake?

👉 A Stream in Snowflake is used to track changes (CDC — Change Data Capture) in a table.

---

## 💡 Simple Idea

👉 Stream = change tracker

It records:
- INSERT
- UPDATE
- DELETE

---

## 🎯 Why Streams?

Without streams:
- You must scan full table ❌

With streams:
- Only process changed data ✅

---

## 🧠 Example

Table:

| id | amount |
|----|--------|
| 1  | 100    |

Update:

UPDATE orders SET amount = 200 WHERE id = 1;

👉 Stream captures:
- Old value (delete)
- New value (insert)

---

# ⚙️ 2. How Streams Work Internally

---

## 📌 Mechanism

- Uses table metadata (micro-partitions)
- Tracks changes without copying data
- Works like a log pointer

---

## 🔄 Flow

Table → Change happens → Stream records → Query Stream → Process

---

## ⚠️ Important

- Stream does NOT store data physically
- It stores offsets to changes

---

# 📊 3. Types of Streams

---

## 🔹 1. Standard Stream (Default)

👉 Tracks all changes:
- INSERT
- UPDATE
- DELETE

Example:

CREATE STREAM orders_stream ON TABLE orders;

---

## 🔹 2. Append-Only Stream

👉 Tracks only INSERTs

Example:

CREATE STREAM orders_stream
ON TABLE orders
APPEND_ONLY = TRUE;

---

## 🔹 3. Insert-Only Stream (External Tables)

👉 Used for:
- External tables (S3 data)

Example:

CREATE STREAM ext_stream
ON EXTERNAL TABLE ext_orders;

---

# 🧠 4. Stream Output Structure

Query Stream:

SELECT * FROM orders_stream;

Output Columns:

- METADATA$ACTION → INSERT / DELETE
- METADATA$ISUPDATE → TRUE/FALSE
- METADATA$ROW_ID → Unique ID

Example Output:

| id | amount | ACTION | ISUPDATE |
|----|--------|--------|----------|
| 1  | 100    | DELETE | TRUE     |
| 1  | 200    | INSERT | TRUE     |

---

# ⚙️ 5. Using Streams with Tasks

Step 1:

CREATE STREAM orders_stream ON TABLE orders;

Step 2:

CREATE TABLE orders_processed (
  id INT,
  amount NUMBER
);

Step 3:

CREATE TASK process_orders
WAREHOUSE = my_wh
SCHEDULE = '1 MINUTE'
AS
INSERT INTO orders_processed
SELECT id, amount
FROM orders_stream;

Step 4:

ALTER TASK process_orders RESUME;

# ⚡ 8. Problems Solved

Without streams:
- Full table scans
- Duplicate processing
- Complex CDC logic

With streams:
- Incremental processing
- Exactly-once
- Built-in CDC

# ⚠️ 9. Important Points

- Stream = CDC tool
- Tracks changes only
- Must be consumed
- Works with Tasks

# 🚨 10. Limitations

- Not long-term storage
- Needs proper consumption
