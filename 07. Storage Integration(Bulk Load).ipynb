# ❄️ Snowflake Storage Integration — Detailed Notes with Examples

---

# 🧠 1. What is Storage Integration in Snowflake?

Storage Integration in Snowflake is a secure, managed connection between Snowflake and external cloud storage like:

- AWS S3
- Azure Blob Storage
- Google Cloud Storage

It allows Snowflake to access data without storing credentials inside Snowflake.

---

## Definition

Storage Integration = IAM-based secure access mechanism to external storage

---

# ⚠️ 2. Problem with Traditional Method

## Traditional Approach

`CREATE STAGE my_stage

URL='s3://mybucket/data/'

CREDENTIALS=(AWS_KEY_ID='xxx' AWS_SECRET_KEY='xxx');`

### Issues

- Credentials stored in SQL
- Security risk
- Hard to rotate keys
- Manual management

---

# ✅ 3. Storage Integration (Modern Approach)

CREATE STORAGE INTEGRATION my_s3_integration

TYPE = EXTERNAL_STAGE

STORAGE_PROVIDER = S3

ENABLED = TRUE

STORAGE_AWS_ROLE_ARN = 'arn:aws:iam::123456789012:role/snowflake-role'

STORAGE_ALLOWED_LOCATIONS = ('s3://mybucket/data/');

---

## Benefits

- No credentials stored
- Uses IAM roles
- Secure and scalable
- Easy maintenance

---

# ⚖️ 4. Comparison

Traditional vs Integration

- Security → Weak vs Strong
- Credentials → Stored vs IAM
- Management → Manual vs Centralized

---

# ☁️ 5. AWS Example

CREATE STORAGE INTEGRATION s3_int

TYPE = EXTERNAL_STAGE

STORAGE_PROVIDER = S3

ENABLED = TRUE

STORAGE_AWS_ROLE_ARN = 'arn:aws:iam::123456789012:role/snowflake-role'

STORAGE_ALLOWED_LOCATIONS = ('s3://mybucket/data/');

---

CREATE STAGE my_stage

URL='s3://mybucket/data/'

STORAGE_INTEGRATION = s3_int;

COPY INTO my_table FROM @my_stage;

---

# ☁️ 6. Azure Example

CREATE STORAGE INTEGRATION azure_int

TYPE = EXTERNAL_STAGE

STORAGE_PROVIDER = AZURE

ENABLED = TRUE

AZURE_TENANT_ID = '<tenant-id>'

STORAGE_ALLOWED_LOCATIONS = ('azure://container/path/');


CREATE STAGE my_stage

URL='azure://container/path/'

STORAGE_INTEGRATION = azure_int;

COPY INTO my_table FROM @my_stage;
