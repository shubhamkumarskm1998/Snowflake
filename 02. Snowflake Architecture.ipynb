# ❄️ Snowflake Architecture — Complete Detailed Notes

---

# 🧠 1. Overview

Snowflake follows a 3-layer architecture:

Storage Layer → Cloud Services Layer → Compute Layer

Key Idea:
- Fully decoupled architecture
- Storage and Compute are independent

---

# 🧱 2. Storage Layer

## What is Storage Layer?

Where all data is stored

### Features

- Uses cloud storage:
  - AWS S3
  - Azure Blob
  - GCP Storage

- Data stored in:
  - Columnar format
  - Compressed format

### Internal Components

- Micro-partitions
- Columnar storage engine
- Metadata tracking

### Micro-Partitions

Data is divided into small chunks (~16MB)

Example:
Table: sales

Partition 1 → rows 1–1000
Partition 2 → rows 1001–2000

### Benefits

- Faster queries
- Automatic partitioning
- No manual indexing

### Important Points

- Fully managed
- Immutable
- Optimized for analytics

---

# 🧠 3. Cloud Services Layer

## What is it?

Brain of Snowflake

### Responsibilities

- Query parsing
- Query optimization
- Metadata management
- Access control & security
- Transaction management

### Internal Components

- Query optimizer
- Metadata store
- Authentication
- Infrastructure manager

Example:

SELECT SUM(revenue) FROM sales;

Flow:
1. Parse query
2. Optimize
3. Send to compute

Important:
- Does NOT store data
- Controls execution

---

# ⚙️ 4. Compute Layer (Virtual Warehouses)

## What is it?

Where queries execute

### Virtual Warehouse

Cluster of compute resources

### Features

- Independent clusters
- Multiple warehouses possible
- Parallel execution

Example:

Teams:
- Engineers → Warehouse A
- Analysts → Warehouse B
- Data Scientists → Warehouse C

### Scaling

#### Scale Up (Vertical)

Increase size

Small → Medium → Large

Use when:
- Queries slow

#### Scale Out (Horizontal)

Add clusters (multi-cluster)

Use when:
- Many users

Important:
- Auto start/suspend
- Pay only when running

---

# 🔄 5. Query Flow

Steps:

1. User sends query
2. Cloud Services parses
3. Compute executes
4. Storage provides data
5. Result returned

Flow:
User → Cloud Services → Compute → Storage → Result

---

# 💰 6. Billing

Two components:

### Storage Cost

- Based on data stored

### Compute Cost

- Based on warehouse size
- Time running

Important:

- Per-second billing
- Auto suspend reduces cost

Example:

10 min usage → pay for 10 min

---

# ⚡ 7. Scaling Summary

Scale Up:
- Bigger machine
- Faster queries

Scale Out:
- More clusters
- More concurrency

Example:

Normal day → Small warehouse
Peak sale → Multi-cluster

---

# ⭐ 8. Key Features

- Separate compute & storage
- Automatic partitioning
- No indexing needed
- High concurrency
- Auto scaling
- Fully managed

---

# 🚨 Important Points

- 3 layers:
  - Storage
  - Cloud Services
  - Compute

- Micro-partitions important
- Virtual warehouse = compute
- Scale up = size increase
- Scale out = cluster increase
- Billing = storage + compute

---

# 🚀 One Line

Snowflake = Decoupled scalable cloud data warehouse

# 🧠 1. Snowflake Account Creation

---

## 💡 What is Snowflake Account?

👉 A Snowflake account is your entry point to use Snowflake platform

- Unique account
- Hosted on cloud (AWS / Azure / GCP)
- Region-specific

---

## 🪜 Steps to Create Snowflake Account

---

### 🔹 Step 1: Go to Snowflake Website

👉 Visit:
https://www.snowflake.com

👉 Click:
Start for Free / Try Snowflake

---

### 🔹 Step 2: Fill Details

Provide:

- Email ID
- Company name
- Region (VERY IMPORTANT)
- Cloud provider:
  - AWS
  - Azure
  - GCP

---

## ⚠️ Important Decision

👉 Choose region & cloud carefully:

- Cannot easily change later
- Impacts latency & cost

---

### 🔹 Step 3: Email Verification

👉 You will receive:

- Activation email
- Login URL
- Temporary credentials

---

### 🔹 Step 4: Login

👉 Open provided URL:

https://<account>.snowflakecomputing.com

👉 Enter:
- Username
- Password

---

### 🔹 Step 5: Initial Setup

After login:

- Change password
- Default role = ACCOUNTADMIN
