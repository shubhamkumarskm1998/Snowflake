# ❄️ Snowflake — Database & Schema (Detailed Notes + Examples)

---

# 🧠 1. What is a Database in Snowflake?

👉 A database is a top-level container that stores all your data

---

## Structure

Account → Database → Schema → Tables → Columns

---

## Key Points

- Organizes data logically
- Contains schemas
- Multiple databases can exist in one account

---

## Example

Company setup:

- sales_db
- hr_db

---

## Important

- Databases are independent
- Can have different permissions

---

# 🗂️ 2. What is a Schema?

👉 A schema is a container inside a database

---

## Structure

Database → Schema → Tables / Views / Functions

---

## Key Points

- Organizes objects
- Logical grouping
- Helps in access control

---

## Example

Inside sales_db:

- raw_data
- processed_data
- analytics

---

# 🧱 3. Objects Inside Schema

- Tables
- Views
- Materialized Views
- Functions
- Stages
- File formats

---

# ⚙️ 4. Creating Database & Schema

## Create Database

CREATE DATABASE sales_db;

## Create Schema

CREATE SCHEMA sales_db.raw_data;

## Use

USE DATABASE sales_db;
USE SCHEMA raw_data;

---

# 🧠 5. Naming Convention

database.schema.table

Example:

sales_db.raw_data.orders

---

# 🔐 6. Access Control

Permissions can be applied at:

- Database
- Schema
- Table

---

# ⚡ 7. Default Schema

- PUBLIC schema

Important:
- Accessible by all
- Avoid for sensitive data

---

# 🔄 8. Database vs Schema

Database → top level
Schema → inside database

---

# 🧠 9. Real Example

sales_db
- raw_data
- processed_data
- analytics

---

# 🚨 Important Points

- Database = container
- Schema = sub-container
- Structure = DB → Schema → Table
- Default = PUBLIC

---

# 🚀 One Line

Database = big container
Schema = folder inside database


# ❄️ Snowflake — Default Schemas in a Database

---

# 🧠 What are Default Schemas?

👉 When you create a database in Snowflake, it automatically creates default schemas inside it.

---

# 📦 Default Schemas in Every Database

When you run:

CREATE DATABASE my_db;

👉 Snowflake automatically creates:

---

## 🟢 1. PUBLIC Schema

---

### 💡 What is PUBLIC Schema?

👉 Default schema available in every database

---

### 📌 Features

- Created automatically
- Accessible to all users (by default)
- Used for:
  - Tables
  - Views
  - Other objects

---

### 🧠 Example

USE DATABASE my_db;
CREATE TABLE PUBLIC.orders (id INT);

👉 Table created in PUBLIC schema

---

### ⚠️ Important Points

- All roles can access it (by default)
- Not recommended for sensitive data
- Good for:
  - Testing
  - Shared data

---

# 🟡 2. INFORMATION_SCHEMA

---

### 💡 What is it?

👉 System schema containing metadata about database objects

---

### 📌 Contains Information About

- Tables
- Columns
- Views
- Schemas
- Constraints

---

### 🧠 Example

SELECT * FROM my_db.INFORMATION_SCHEMA.TABLES;

---

### 🎯 Use Case

- Metadata queries
- Auditing
- Data discovery

---

### ⚠️ Important Points

- Read-only schema
- Cannot create tables here
- Used internally by Snowflake

---

# 🔄 Summary

---

| Schema | Purpose | Access |
|--------|--------|--------|
| PUBLIC | Default working schema | Open (by default) |
| INFORMATION_SCHEMA | Metadata | Read-only |

---

# 🚨 Important Points (Exam / Interview)

---

- Every database has 2 default schemas
- PUBLIC → user objects
- INFORMATION_SCHEMA → metadata
- PUBLIC is open by default (be careful)

---

# 🚀 One Line Revision

👉 Snowflake DB = PUBLIC (data) + INFORMATION_SCHEMA (metadata)


🔄 Execution Flow

Set role → permissions

Set warehouse → compute

Create database → container

Create schema → organize data

# ❄️ Snowflake — Table Types (Detailed Notes + Examples)

---

# 🧠 1. Overview

Snowflake provides different table types based on:

- Data persistence
- Cost
- Use case

---

# 📊 Types of Tables

1. Permanent Tables
2. Transient Tables
3. Temporary Tables
4. External Tables
5. Iceberg Tables

---

# 🟢 2. Permanent Tables

## Features
- Default table type
- Supports Time Travel + Fail-safe
- Highly durable

## Example

CREATE TABLE orders (
    id INT,
    amount NUMBER
);

INSERT INTO orders VALUES (1, 100);

SELECT * FROM orders;

## Use Case
- Production data
- Critical data

---

# 🟡 3. Transient Tables

## Features
- No Fail-safe
- Lower cost
- Limited recovery

## Example

CREATE TRANSIENT TABLE staging_orders (
    id INT,
    amount NUMBER
);

INSERT INTO staging_orders VALUES (2, 200);

SELECT * FROM staging_orders;

## Use Case
- ETL staging
- Intermediate data

---

# 🔵 4. Temporary Tables

## Features
- Session-based
- Auto deleted
- Not visible to others

## Example

CREATE TEMP TABLE temp_orders (
    id INT,
    amount NUMBER
);

INSERT INTO temp_orders VALUES (3, 300);

SELECT * FROM temp_orders;

## Use Case
- Debugging
- Temporary calculations

---

# 🟣 5. External Tables

## Features
- Data stored outside Snowflake
- Query directly

## Example

CREATE EXTERNAL TABLE ext_orders
WITH LOCATION = @my_stage
FILE_FORMAT = (TYPE = 'CSV');

SELECT * FROM ext_orders;

## Use Case
- Data lake
- Large datasets

---

# 🧊 6. Iceberg Tables

## Features
- Open table format
- Works with data lake
- ACID support

## Example (conceptual)

CREATE ICEBERG TABLE iceberg_orders (
    id INT,
    amount NUMBER
);

## Use Case
- Lakehouse
- Multi-engine access

---

# 🔄 Comparison

| Feature | Permanent | Transient | Temporary |
|--------|----------|----------|-----------|
| Persistence | Yes | Yes | No |
| Fail-safe | Yes | No | No |
| Scope | Global | Global | Session |

---

# 🧠 Real ETL Example

Raw Data → External Table
Staging → Transient Table
Final → Permanent Table

---

# 🚀 One Line

Permanent = safe
Transient = cheaper
Temporary = session
External = outside
Iceberg = lakehouse


# ❄️ Snowflake — Retention Time & Fail-safe (Detailed Explanation)

---

# 🧠 1. What is Retention Time (Time Travel)?

👉 Retention Time = how long Snowflake keeps historical data for recovery

This feature is called:

👉 Time Travel

---

## 💡 What Time Travel Allows You To Do

- Recover deleted data
- Undo updates
- Query past versions of data

---

## 📌 Example

-- Create table
CREATE TABLE orders (id INT, amount NUMBER);

-- Insert data
INSERT INTO orders VALUES (1, 100);

-- Delete data
DELETE FROM orders;

-- Recover using Time Travel
SELECT * FROM orders AT (OFFSET => -60*5);

---

## ⏱️ Retention Period

Permanent → 1–90 days
Transient → 0–1 day
Temporary → No Time Travel

---

## ⚠️ Important

- More retention = more cost
- Configurable

---

# 🔐 2. What is Fail-safe?

👉 Fail-safe = last backup layer managed by Snowflake

---

## 💡 Key Idea

After Time Travel expires:

👉 Data still available for 7 days

---

## 📌 Features

- Fixed: 7 days
- Not configurable
- Requires Snowflake support

---

## 🧠 Example

Delete → Time Travel expired → Still recoverable via support

---

# 🔄 3. Timeline

Active Data

↓

Time Travel(0-90 days)

↓

Fail-safe(7 days)

↓

Deleted

---

# 🔄 4. Difference

Time Travel → user recovery
Fail-safe → Snowflake recovery

---

# 🚨 Important Points

- Permanent → both
- Transient → no fail-safe
- Temporary → none

---

# 🚀 One Line

Time Travel = user undo
Fail-safe = backup layer


# ❄️ Snowflake — View Types (Detailed Notes + Examples)

---

# 🧠 1. What is a View?

👉 A view is a virtual table based on a SQL query

- Does NOT store data (mostly)
- Stores only the query logic
- Fetches data from underlying tables

---

## 💡 Simple Idea

👉 View = Saved SQL query

---

## 🧠 Example

CREATE VIEW sales_view AS
SELECT id, amount FROM orders;

SELECT * FROM sales_view;

---

# 📊 2. Types of Views

1. Standard View
2. Secure View
3. Materialized View

---

# 🟢 3. Standard View

## Features

- Stores only query
- Always shows latest data
- No storage cost

## Example

CREATE VIEW high_value_orders AS
SELECT * FROM orders WHERE amount > 1000;

## Use Case

- Simplify queries
- Reuse logic

---

# 🔐 4. Secure View

## Features

- Hides query logic
- Prevents data leakage
- Used for sharing

## Example

CREATE SECURE VIEW secure_orders AS
SELECT id, amount FROM orders;

## Use Case

- Sensitive data
- Data sharing

---

# 🟡 5. Materialized View

## Features

- Stores data physically
- Faster queries
- Auto refresh

## Example

CREATE MATERIALIZED VIEW total_sales AS
SELECT region, SUM(amount)
FROM orders
GROUP BY region;

## Use Case

- Aggregations
- Performance optimization

---

# 🔄 Comparison

| Feature | Standard | Secure | Materialized |
|--------|----------|--------|--------------|
| Stores Data | No | No | Yes |
| Performance | Medium | Medium | Fast |
| Security | Normal | High | Normal |

---

# 🧠 Real Example

CREATE TABLE orders (
    id INT,
    region STRING,
    amount NUMBER
);

CREATE VIEW us_orders AS
SELECT * FROM orders WHERE region = 'US';

CREATE SECURE VIEW public_orders AS
SELECT id, amount FROM orders;

CREATE MATERIALIZED VIEW sales_summary AS
SELECT region, SUM(amount)
FROM orders
GROUP BY region;

---

# 🚨 Important Points

- View = virtual
- Secure = protected
- Materialized = stored

---

# 🚀 One Line

View = query
Secure = secure query
Materialized = stored result
